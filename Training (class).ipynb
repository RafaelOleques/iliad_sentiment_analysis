```python
from trainer import Trainer
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
import matplotlib.pyplot as plt
```

```python
path = "./data/iliad_sentiments.csv"
target = "class"
```

# Standard class

## Without lemmatization

```python
trainer_nb = Trainer(path, target)
trainer_dt = Trainer(path, target)
trainer_knn = Trainer(path, target)

nro_folds = 4
lemmatize = False

trainer_nb.cross_validation(model=MultinomialNB(), nro_folds=nro_folds, lemmatize=lemmatize, ohe=False)
trainer_dt.cross_validation(model=DecisionTreeClassifier(), nro_folds=nro_folds, lemmatize=lemmatize, ohe=True)
trainer_knn.cross_validation(model=KNeighborsClassifier(), nro_folds=nro_folds, lemmatize=lemmatize, ohe=True)
```

**Accuracy**

```python
results=[]
results.append(trainer_nb.get_metric("accuracy"))
results.append(trainer_dt.get_metric("accuracy"))
results.append(trainer_knn.get_metric("accuracy"))

plt.boxplot(results, labels=["NB", 'DT','KNN'], showmeans=True)
plt.title("Accuracy: standard attributes")
plt.show()
```

**Recall**

```python
results=[]
results.append(trainer_nb.get_metric("recall"))
results.append(trainer_dt.get_metric("recall"))
results.append(trainer_knn.get_metric("recall"))

plt.title("Recall: standard attributes")
plt.boxplot(results, labels=["NB", 'DT','KNN'], showmeans=True)
plt.show()
```

**Precision**

```python
results=[]
results.append(trainer_nb.get_metric("precision"))
results.append(trainer_dt.get_metric("precision"))
results.append(trainer_knn.get_metric("precision"))

plt.boxplot(results, labels=["NB", 'DT','KNN'], showmeans=True)
plt.title("Precision: standard attributes")
plt.show()
```
